```
import geopandas as gp
import pandas as pd 
import numpy as np
import folium
import matplotlib.pyplot as plt

```

```
cairo_raw=gp.read_file('/content/stanford-vn895fq9113-geojson.json')
```

```
cairo_raw.head()
```

```
cairo_raw.columns
```

```
# only Great cairo regions, Cairo+Giza
cairo_raw=cairo_raw[(cairo_raw['name_1']=='Al Qahirah')| (cairo_raw['name_1']=='Al Jizah')]

cols=['id', 'id_0', 'iso', 'name_0', 'id_1', 'name_1', 'id_2', 
       'hasc_2', 'ccn_2', 'cca_2', 'type_2', 'engtype_2', 'nl_name_2',
       'varname_2']
# drop unneccessary columns
cairo_raw.drop(cols,axis=1,inplace=True)
cairo_raw.rename(columns={'name_2':'Neighborhood'},inplace=True)
cairo_raw.reset_index(inplace=True)
cairo_raw.head()
```

```
cairo_raw.drop('index',axis=1, inplace=True)

cairo_raw
```

```
print(cairo_raw.geometry.centroid.x,cairo_raw.geometry.centroid.y)
```

```
cairo_raw['lat']=cairo_raw.geometry.centroid.y
cairo_raw['long']=cairo_raw.geometry.centroid.x

cairo_raw.head()
```

```
cairo=cairo_raw[['Neighborhood','lat','long']]
cairo.head()
```

```
cairo_map=folium.Map(location=[30.0444, 31.2357],zoom_start=11)
cairo_map
```

```
for lat, lng, neig in zip(cairo['lat'], cairo['long'], cairo['Neighborhood']):
    #label = '{}, {}'.format(neig, borog)
    label = folium.Popup(neig, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(cairo_map)  
    
cairo_map
```

```
cairo.to_csv('cairo.csv')
```
